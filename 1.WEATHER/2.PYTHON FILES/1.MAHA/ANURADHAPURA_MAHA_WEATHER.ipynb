```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load your weather data into a DataFrame
# Replace 'AMPARA_MAHA_SEASON.csv' with the path to your actual data file
data = pd.read_csv('ANURADHAPURA_MAHA_SEASON.csv')
```

```python
# Display the first few rows of the DataFrame to understand its structure
print(data.head())
```

```python
# Generate descriptive statistics for the weather data
stats_summary = data.describe()
print("\nStatistical Summary:")
print(stats_summary)
```

```python
# Set up the matplotlib figure for histograms and boxplots
plt.figure(figsize=(15, 10))

# List of columns to plot
columns = ['Maha Total Rainfall', 'Maha Temp Average', 'Maha Average Of Max Temp', 'Maha Highest Monthly  Average of MaxTemp','Maha Average Of Min Temp','Maha Lowest Monthly  Average of Min Temp']
labels = ['Total Rainfall (mm)', 'Average Temperature (°C)', 'Average Max Temperature (°C)','Higest of Average monthly Max Temperature (°C)' ,'Average Min Temperature (°C)','Lowest of Average monthly Min Temperature (°C)']

# Create subplots for histograms and boxplots
fig, axs = plt.subplots(nrows=6, ncols=2, figsize=(15, 20))

# Loop through each column to create histograms and boxplots
for i, (col, label) in enumerate(zip(columns, labels)):
    # Histogram with KDE
    sns.histplot(data[col], kde=True, bins=30, ax=axs[i, 0])
    axs[i, 0].set_title(f'Distribution of {label}')
    axs[i, 0].set_xlabel(label)
    axs[i, 0].set_ylabel('Frequency')

    # Boxplot
    sns.boxplot(y=data[col], ax=axs[i, 1])
    axs[i, 1].set_title(f'Boxplot of {label}')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

```python
# Create subplots
fig, axs = plt.subplots(nrows=6, ncols=1, figsize=(15, 30))

# Loop through each column and plot in separate subplots
for ax, col, label in zip(axs, columns, labels):
    ax.plot(data['Year'], data[col], marker='o', label=label)
    ax.set_title(label)
    ax.set_xlabel('YEAR')
    ax.set_ylabel(label)
    ax.grid()
    ax.legend()

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the plots
plt.show
```
